### 4. Array Shape, reshape and transpose

```python
import numpy as np
```

```python
arr = np.array([['Shirts','Hoodies','Tshirt'],['Tops','Long Frocks','Short Kurthis']])
```

```python
arr.shape
```

```python
# reshape() : This will change the dimentions of the numpy array.
arr.reshape(3,2)
```

```python
arr.reshape(1,6)
```

```python
arr.reshape(1,3,2)
```

```python
arr.reshape(6)
```

```python
# ravel() : Flatten numpy array to 1D array.
arr.ravel()
```

```python
# transpose of matrix : rows will convert to columns and columns will convert into rows.
arr # Original Array
```

```python
arr.T # Transpose of original array.
```

```python
# creating high Dimentional Array quickly.

arr_5D = np.arange(1,49).reshape(2,2,3,2,2) 
# On multiplying dimentions we can know the no of elements in n array.
arr_5D
```
