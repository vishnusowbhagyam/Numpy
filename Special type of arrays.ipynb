### 3. Special type of numpy array

```python
import numpy as np
```

```python
# Numpy array which initialize with Zero

# 1D zero array
zero_arr = np.zeros(5)
zero_arr
```

```python
# 2D zero array
zero_arr = np.zeros((2,3))
zero_arr
```

```python
# 3D zero array
zero_arr = np.zeros(((2,2,3)))
zero_arr
```

```python
# creating numpy array which are initialize with ones.

# 1D one's array

ones_arr = np.ones(5)
ones_arr
```

```python
# 2D one's array

ones_arr = np.ones((3,4))
ones_arr
```

```python
# 3D one's array

ones_arr = np.ones(((2,3,5)))
ones_arr
```

```python
# creating numpy array with specific value by using full method

# 1D ARRAY FILLED WITH 90
arr = np.full(5,90)
arr
```

```python
# 2D ARRAY FILLED WITH 10
arr = np.full((3,5),10)
arr
```

```python
# 3D ARRAY FILLED WITH 9
arr = np.full(((3,2,5)),9)
arr
```

```python
# Identity matrix : Having diagonal values with ones.

identity = np.eye(4)
identity
```

### random() : It is used to generate random elements

```python
# 1D ARRAY OF RANDOM ELEMENTS
# RAND() : GENERATES RANDOM ELEMENTS BETWEEN O AND 1
# RAND() FUNCTION USING UNIFORM DISTRIBUTION WHILE SELECTING THE VALUES

np.random.rand(5)
```

```python
# 2D array of random elements (0 to 1) using rand()
np.random.rand(2,3)
```

```python
# 3D array of random elements (0 to 1) using rand()
np.random.rand(2,2,4)
```

```python
# random() : random works same as rand() function
# random() function receives the input in tuple format
np.random.random(10)
```

```python
np.random.random((2,3))
```

```python
# randn() : Generates most of values in between -3 to +3
# Here n stands for standard normal distribution (mean = 0, sd = 1)

#1D
np.random.randn(15)
```

```python
# 2D
np.random.randn(2,3)
```

```python
# randint() : It generates random integer elements between a range
rand_int = np.random.randint(1,10,15) 
# (a,b,c) : a is start and inclusive, b is end and exclusive and c is count
rand_int
```

```python
# 2D 
rand_int = np.random.randint(1,10,(2,6))
rand_int
```

```python
# choice() : It works specifically on catagorical data.
# choice() function is used for selecting random elements from categorical list

arr = np.array(['raj','vishnu','venkat','krishna'])
np.random.choice(arr,size = 2)
```

```python
# It wont give the same names repeatedly
np.random.choice(arr,size = 2,replace = False)
```

```python
np.random.choice(arr,size = 10)
```

```python
# giving the probability of getting every name
np.random.choice(arr,size = 10,p=[0.1,0.2,0.5,0.2])
```

```python
# arange() : creates a sequence of elements
arr = np.arange(1,10,2) # arange(start,end,step)
arr
```

```python
# shuffle() : re-arranges the elements
# shuffle() function rearrange the elements in the same array.
np.random.shuffle(arr)
arr
```

```python
# permutation() function returns new shuffled array
arr2 = np.random.permutation(arr)
arr2
```

```python
#linspace() : linearly spaced numpy array
arr3 = np.linspace(1,10,5)
arr3
```
