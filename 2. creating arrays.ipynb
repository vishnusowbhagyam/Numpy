### 2. Creating numpy array(1D,2D,3D......)

```python
import numpy as np
```

```python
# 1D numpy array

arr1 = np.array([1,2,3,4,5])
print(arr1)

# checking type of array
print(type(arr1))

# datatype of numpy array elements
print(type(arr1.dtype))
```

```python
# checking dimentions of an array

arr1.ndim
```

```python
# float elements into numpy array
# if we insert one float value into array than the entire array converts into float values.

arr1 = np.array([1,2,3,4,5.5])
print(arr1)
print(type(arr1))
print(arr1.dtype)
```

```python
arr1 = np.array([1,2,3,'hello',5.5])
print(arr1)
print(type(arr1))
print(arr1.dtype)

# U32 : unicode string type with 32 length
```

```python
# 2D numpy array

arr2 = np.array([['Bahubali','Devara','Kanthara'],['Prabhas','NTR','Rishab Shetty']])
print(arr2)
print(arr2.ndim)
```

```python
# To check the shape of an array 
arr2.shape
```

```python
arr2 = np.array([[1,2,3],[4,5,6]])
arr2
```

```python
# 3D numpy array : we can create 3D array by using 2d arrays.
arr3 = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])
arr3
                 
```

```python
arr3.ndim
```

```python
arr3.shape
```

```python
# 4D NUMPY ARRAY
arr4 = np.array([[[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]],[[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]]])
arr4
                
```

```python
arr4.ndim
```

```python
arr4.shape
```
