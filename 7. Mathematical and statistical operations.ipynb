### Mathematical and statistical operations

```python
import numpy as np
```

```python
arr = np.array([[1,2,3],[4,5,6]])

print(arr + 10) # adding 10 to every element in an array.
```

```python
# multiplying 2 to every element in an array.
print(arr * 2)
```

```python
print(arr ** 2)
```

```python
print(arr - 1)
```

```python
salary = np.array([10,15,25,30,35,50,80,45,60,75,100,155])

# sum of all the salaries
np.sum(salary)
```

```python
# finding minimum salary
np.min(salary)
```

```python
# max of salary
np.max(salary)
```

```python
# average of salary
np.mean(salary)
```

```python
# variance of a salary
np.var(salary)
```

```python
# standard deviation
np.std(salary)
```

```python
# finding percentile of a salary at 75 percentage
np.percentile(salary,75)
```

```python
# finding percentile of salary in decimal percentage value
np.percentile(salary,0.75)
```

```python
# argmin() : finding the minimum value index
np.argmin(salary)
```

```python
# argmax() : finding max value index position
np.argmax(salary)
```

```python
# finding log value with base 10
np.log10(1000)
```

```python
# finding log value with base 2
np.log2(20)
```

```python
# finding log value with constant base e = 2.71
np.log(50)
```

```python
# finding square root of a value
np.sqrt(300)
```

```python
# rounding the float value upto n positions
pi = 3.1752
np.round(pi)
```

```python
np.floor(pi)
```

```python
np.ceil(pi)
```

```python
# finding absolute value which means giving positive value for any value
np.abs(-23)
```

```python
fruits = np.array(['apple','banana','grapes','orange','mango'])
np.sort(fruits) # sorting in a lexographical order.
```

```python
fruits_price = np.array([150,50,80,100,200])
np.sort(fruits_price)
```

```python
# sorting the prices and reversing it.
np.sort(fruits_price)[::-1]
```

```python
arr = np.array([[10,5,15],[20,35,15]])
np.sort(arr)
```
