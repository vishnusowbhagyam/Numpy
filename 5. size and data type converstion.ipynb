### Size and data type convertion

```python
import numpy as np
```

```python
# size: number of elements in the numpy array
songs = np.array(['Monica','Fear song','Peelings'])
songs.size
```

```python
songs.dtype
```

```python
song_duration = np.array([3.58,3.2,5.09])
int_arr = song_duration.astype(int) # converting float data type into integer.
int_arr
```

```python
# string to int converstion
song_earning = np.array(['500','300','1000'])

song_earning.dtype
```

```python
int_arr = song_earning.astype(int) # converting string into integer
int_arr
```

```python
float_arr = song_earning.astype(float)
float_arr
```

```python
np.array([True,False,True]).astype(int) # converting boolean values into integers.
```

```python
np.array([0,-1,4,5]).astype(bool) # converting integers into boolean.
```
